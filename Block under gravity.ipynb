```python
%matplotlib inline
from matplotlib import animation, rc
from IPython.display import HTML
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def gravity():
    global x
    global y
    global vel
    global grid_size
    global g
    global no_particles
    
    for i in range(no_particles):
        if y[i] <= 0:
            vel[i] = -vel[i]/1.5
    
        else:
            vel[i] = vel[i] + g * timestep
    
        y[i] = y[i] + vel[i] * timestep
    
    return
    
```

```python
grid_size = 5
no_particles = 1
x = []
y = []
vel = []
timestep = 0.01
g = -98

for i in range(no_particles):
    #x.append(np.random.rand() * grid_size)
    #y.append(np.random.rand() * grid_size)
    x.append(grid_size/2)
    y.append(2 * grid_size)
    vel.append(np.random.rand() * -100)

x1 = x
x2 = x
x3 = x
x4 = x

y1 = y
y2 = y
y3 = y
y4 = y

fig = plt.figure()
ax = plt.axes(xlim=(0, grid_size*2), ylim=(0, grid_size*2), aspect='equal')
```

```python
line1, = ax.plot([], [], 'ks', alpha = 1, markersize = 20)

line2, = ax.plot([], [], 'ks', markersize = 15 )

line3, = ax.plot([], [], 'ks', markersize = 10 )
line4, = ax.plot([], [], 'ks', markersize = 5 )
line5, = ax.plot([], [], 'ks', markersize = 2 )

lines = [line1, line2, line3, line4, line5]
```

```python
def init():
    line1.set_data([], [])
    line2.set_data([], [])
    line3.set_data([], [])
    line4.set_data([], [])
    line5.set_data([], [])
    return lines
```

```python
def animate(i):
    x1 = x
    y1 = y
    
    x2 = x1
    y2 = y1
    
    x3 = x2
    y3 = y2
    
    x4 = x3
    y4 = y3
    
    gravity()
    line1.set_data(x, y)
    line2.set_data(x1, y1)
    line3.set_data(x2, y2)
    line4.set_data(x3, y3)
    line5.set_data(x4, y4)
    fig.canvas.draw()
    return line1,
```

```python
anim = animation.FuncAnimation(fig, animate, init_func = init, frames = 400, interval = 30, blit=True)
plt.show()
HTML(anim.to_html5_video())
```
