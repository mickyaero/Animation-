```python
%matplotlib inline
from matplotlib import animation
from IPython.display import HTML
import numpy as np
import matplotlib.pyplot as plt

def floor_update():
    global x
    global y
    global timestep
    global time
    
    for i in range(lg):
        y2[i] = np.sin(grid_size/4 * 0.785 * time + x2[i])
    time = time + timestep
    return

fig, ax = plt.subplots()
ax.set_xlim([0,10])
ax.set_ylim([-5,5])

grid_size = 5
lg = 50
timestep = 0.01
time = 0
x2 = np.linspace(0, grid_size * 2, lg)
y2 = []
for i in range(lg):
    y2.append(grid_size/4)   


collection = plt.fill_between(x2, y2, -5, alpha = 0.8)

def animate(i):
    floor_update()
    path = collection.get_paths()[0]
    for i in range(lg):
        path.vertices[i][1] = y2[i]

anim = animation.FuncAnimation(fig, animate,
                        frames=125, interval=30)
plt.show()
HTML(anim.to_html5_video())
```
